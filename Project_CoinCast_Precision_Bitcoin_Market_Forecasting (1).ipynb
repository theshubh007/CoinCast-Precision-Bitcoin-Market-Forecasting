```python
# Importing MinMaxScaler from sklearn.preprocessing module
from sklearn.preprocessing import MinMaxScaler

# Creating a MinMaxScaler object with feature range scaled between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))

# Reshaping the closing price values into a 2D array and scaling the data
scaled_data = scaler.fit_transform(np.array(df_close.values).reshape(-1, 1))
```

## Splitting the Data

```python
# Splitting the data into training and testing sets
# 75% of the data for training, 25% for testing
train_size = int(len(df_close) * 0.75)
test_size = len(df_close) - train_size

# Printing the sizes of the training and testing sets
print("Train Size:", train_size, "Test Size:", test_size)

# Extracting the training and testing data from the scaled data
# For training data, select the first 'train_size' elements
train_data = scaled_data[:train_size, 0:1]
# For testing data, select 'test_size' elements starting from 'train_size - 60'
test_data = scaled_data[train_size - 60:, 0:1]
```

```python
# Checking the shapes of the training and testing data arrays
# This provides information about the dimensions of the arrays
train_data.shape, test_data.shape
```

## Creating training set

```python
# Creating a Training set with 60 time-steps and 1 output
x_train = []  # List to store input sequences
y_train = []  # List to store output values

# Iterating over the training data to create input-output pairs
# Each input sequence contains 60 time-steps, and the corresponding output is the next time-step value
for i in range(60, len(train_data)):
    # Extracting input sequence of length 60 and appending it to x_train
    x_train.append(train_data[i - 60:i, 0])
    # Extracting the output value (next time-step) and appending it to y_train
    y_train.append(train_data[i, 0])
```

```python
# Convert to numpy array
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshaping the input
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

## MODEL 1: ARIMA

model name: arima_model

```python
# Import necessary libraries
from statsmodels.tsa.arima.model import ARIMA

# Load your dataset into a DataFrame (df_close should be the "Close" price series)
# Example: df_close = pd.read_csv('your_file.csv')['Close']
df_close1 = df_close.sort_index()

# Splitting the data into training and testing sets
train_size = int(len(df_close1) * 0.8)  # 80% training, 20% testing
train_data = df_close1[:train_size]
test_data = df_close1[train_size:]

# Train ARIMA model
order = (1, 2, 1)  # Adjust order based on ACF/PACF plots or auto_arima
arima_model = ARIMA(train_data, order=order)
arima_fit = arima_model.fit()

# Generate predictions for the test set
start = len(train_data)
end = len(train_data) + len(test_data) - 1
predictions = arima_fit.predict(start=start, end=end, typ='levels')

# Calculate evaluation metrics
RMSE = np.sqrt(mean_squared_error(test_data, predictions)).round(2)
MAE = mean_absolute_error(test_data, predictions).round(2)
MSE = mean_squared_error(test_data, predictions).round(2)
R2 = round(r2_score(test_data, predictions), 2)

print("\nARIMA Model Evaluation Metrics:")
print(f"Root Mean Squared Error (RMSE): {RMSE}")
print(f"Mean Absolute Error (MAE): {MAE}")
print(f"Mean Squared Error (MSE): {MSE}")
print(f"R-squared (R2): {R2}")

# Forecast the next 60 days
print("\nForecasting the next 60 days...")
forecast = arima_fit.get_forecast(steps=60)
forecasted_mean = forecast.predicted_mean
forecast_index = pd.date_range(start=df_close1.index[-1], periods=61, freq='D')[1:]  # Next 60 days

# Combine historical and forecasted data for visualization
combined_index = train_data.index.append(test_data.index).append(forecast_index)
combined_data = pd.concat([train_data, test_data, forecasted_mean], ignore_index=True)

# Visualize the results
plt.figure(figsize=(12, 6))
plt.plot(combined_index[:len(train_data)], train_data, label="Train Data", color="blue")
plt.plot(combined_index[len(train_data):len(train_data) + len(test_data)], test_data, label="Test Data", color="green")
plt.plot(combined_index[len(train_data):], combined_data[len(train_data):], label="ARIMA Predictions", color="orange")
plt.title("ARIMA Predictions vs Actual Prices")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.show()
```
